Task 1 : Prediction using Supervised ML

```python
# reading the data
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
df.head()
```

```python
# general information about the dataset
df.info()
```

```python
# numerical statistics about the dataset
df.describe()
```

```python
# Scatter Plot to better understand the data
import matplotlib.pyplot as plt
plt.scatter(df['Hours'], df['Scores'])
plt.title("No. of Study Hours V/S Obtained Scores")
plt.xlabel("No. of Study Hours")
plt.ylabel("Obtained Scores")
plt.show()
```

```python
# creating a simple linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# fitting our dataset in the model
model.fit(df[['Hours']], df['Scores'])
```

```python
# Performance evaluation of the Simple LR Model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(df[['Hours']])
mean_abs = mean_absolute_error(df['Scores'], y_pred)
mean_sq = mean_squared_error(df['Scores'], y_pred)
r_sq = r2_score(df['Scores'], y_pred)
print("Mean Absolute Error = ", mean_abs)
print("Mean Sqaured Error = ", mean_sq)
print("R2 Score = ", r_sq)
```

```python
# Prediction using the Simple LR Model
y_pred = model.predict([[9.25]])
print("No. of Hours = ", 9.25)
print("Predicted Score = ", y_pred[0])
```
